```python
import matplotlib.pyplot as plt
```

```python
nickst = {
    "figure.facecolor": "#151320",
    "text.color": "#FFFF80",
    "font.size": 12,
    "axes.facecolor": "#22212C",
    "axes.labelcolor": "#FFFF80",
    "axes.labelsize": 14,
    "axes.titlesize": 16,
    "axes.grid": True,
    "grid.alpha": 0.4,
    "grid.color": "#373252",
    "grid.linestyle": "-",
    "xtick.color": "#A95B86",
    "ytick.color": "#7FFBE6"
}
plt.rcParams.update(nickst)
```

```python
from random import randint as rn

def plot_lines():
    plt.plot(
        [x for x in range(0, 100)],
        [(x**2)/10 for x in range(0, 100)]
    )
    plt.plot(
        [x for x in range(0, 100)],
        [(x**2)/15 for x in range(100, 0, -1)]
    )
    plt.plot(
        [x for x in range(0, 100)],
        [x*5 for x in range(0, 100)]
    )
    plt.plot(
        [x for x in range(0, 100)],
        [x*10 for x in range(100, 0, -1)]
    )
    plt.xlabel("X axis")
    plt.ylabel("Y axis")

def plot_bars():
    plt.bar(
        [x for x in range(0, 10)],
        [rn(1, 10) for _ in range(0, 10)]
    )
    plt.grid(False)
    plt.xlabel("X axis")
    plt.ylabel("Y axis")

plot_lines()
plt.show()
plot_bars()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
plt.title("Sample line graph")
plot_lines()
plt.subplot(2, 2, 2)
plt.title("Random histogram")
plot_bars()
plt.subplot(2, 2, 3)
plt.title("Random histogram again")
plot_bars()
plt.subplot(2, 2, 4)
plt.title("Line graph again!")
plot_lines()
plt.subplots_adjust(left=0.1, right=0.9, top=1.0, bottom=0.1, wspace=0.4, hspace=0.8)
```

```python
with open("nickst.mplstyle", "w") as file:
    for key, value in nickst.items():
        if str(value)[0] == "#": value = "\"" + value + "\""
        file.write(f"{key} : {value}\n")
```

```python
plt.style.use("./nickst.mplstyle")
```
